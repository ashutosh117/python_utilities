```python
import pandas as pd
import urllib2
from bs4 import BeautifulSoup
```

```python
wiki = "https://en.wikipedia.org/wiki/List_of_state_and_union_territory_capitals_in_India"
page = urllib2.urlopen(wiki)
```

```python
#save the page in beautifulsoup format
soup = BeautifulSoup(page)
#print soup.prettify() #to look at the html page in nested format
```

```python
soup.title.string 
```

```python
all_tables = soup.find_all('table')
right_table = soup.find_all('table',{"class" : 'wikitable sortable plainrowheaders'})[0]
#right_table
```

```python
A,B,C,D,E,F=[],[],[],[],[],[]
for row in right_table.find_all('tr'):
    cells = row.find_all('td')
    if len(cells) == 6:
        state = row.find_all('th')
        A.append(cells[1].find(text = True))
        B.append(state[0].find(text = True))
        C.append(cells[2].find(text = True))
        D.append(cells[3].find(text = True))
        E.append(cells[4].find(text = True))
        F.append(cells[5].find(text = True))
```

```python
df = pd.DataFrame(B,columns = ['State/UT'])
df['Admin_cap'] = A
df['Judiciary_cap'] = C
df['legislative_cap'] = D
df['year_cap'] = E
df['former_cap'] = F
df
```
