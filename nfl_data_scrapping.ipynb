```python
#scrapping NFL data using python 2 and BeautifulSoup 
import os
from bs4 import BeautifulSoup
import urllib2
from urllib import urlretrieve
```

```python
urllist=['madden-nfl-2003.html', 'madden-nfl-2004.html', 'madden-nfl-2005.html', 'madden-nfl-06.html', 
'madden-nfl-07.html', 'madden-nfl-08.html', 
'madden-nfl-09.html', 'madden-nfl-10.html', 'madden-nfl-11.html', 'madden-nfl-12.html']

urlorig = 'http://maddenratings.weebly.com/'
output_dir = '/home/t1/data_science/data_analysis_r/nfl_data_analysis/files/data/'
```

```python
for url in urllist:
    u = urlorig + url
    html = urllib2.urlopen(u)
    soup = BeautifulSoup(html,'html.parser')
    links = soup.find_all('a')
    for link in links:
        if link.has_attr('href'):
            href = link.get('href')
        else:
            continue
        #print href
        if href[-4:] not in ['.csv','.xls','xlsx']:
            continue
        href = urlorig+href
        #print href.split('/')[-1]
        filename = os.path.join(output_dir,href.split('/')[-1])
        urlretrieve(href,filename)
        print 'downloading file %s...' %(href.split('/')[-1])
```
